Lista que almacena las duraciones de las peliculas

```python
duracion_peliculas = df_movies['duration'].to_list()
len(duracion_peliculas)
```

Funcion para iterar sobre la lista, transformar cada string a numero decimal y devolver el valor numerico mas alto

```python
def maxima_duracion(lista):
  duration = [float(d) for d in lista]
  return max(duration)

maxima_duracion(duracion_peliculas)
```

**¿Cual es la duracion media de las peliculas por decada en Netflix?**

```python
df_movies.head()
```

```python
df_movies.dtypes
```

```python
df_movies['duration'] = df_movies['duration'].astype(float)
df_movies.head()
```

```python
df_movies.dtypes
```

Nueva columna con el valor de la decada correspondiente a cada año

```python
df_movies['decade'] = (df_movies['release_year'] // 10) * 10 # // division entera
df_movies.head()
```

Agrupar por decada y duracion promedio de peliculas por decada

```python
duration = round(df_movies.groupby('decade')['duration'].mean(), 2) # .idxmax()
duration
```
